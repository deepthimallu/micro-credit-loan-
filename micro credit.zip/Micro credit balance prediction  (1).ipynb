### Skewness Reduction Techinque

### Power Transformer

```python
#Treating of skewness using Powertransform function.

from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson')
x=pt.fit_transform(x)
x   
```

### Scailing

```python
# Scaling the data using MinMaxScaler.
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
x
```

```python
#lets use oversampling technique on minority class using smote method
#importing smote from imblearn
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x_over,y_over=sm.fit_resample(x,y)
y_over.value_counts()
```

```python
#Count plot
sns.countplot(y_over)
```

### Principle Component Analysis

```python
#importing PCA
from sklearn.decomposition import PCA
pca=PCA()
pca.fit_transform(x_over)
var_cumu=np.cumsum(pca.explained_variance_ratio_)*100
var_cumu
```

```python
k=np.argmax(var_cumu>99)
k
```

```python
#ploting cumulative variance
plt.ylabel('Cumulative Explained varience')
plt.xlabel('Principle components')
plt.axvline(x=k, color='k',linestyle='--')
plt.axhline(y=99, color='r',linestyle='--')

plt.plot(var_cumu)
plt.show()
```

Principle component analysis gives the relation between principle components and cumulative explained varience.From Principle component analysis we can know the best features which mostly contribute for our prediction.

### Train Test Split

```python
#importing train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_over,y_over,test_size=.30,random_state=40)
```

### Classification Models

```python
#Importing some classifiers
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
#lets import some of the evaluation metrics for classification problem
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,roc_curve,roc_auc_score
#importing grid search cv for hyper parameter tuning and cross validation score for testing our model
from sklearn.model_selection import GridSearchCV,cross_val_score
```

### Logistic Regression

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred_lr=lr.predict(x_test)
print('Training score:',lr.score(x_train,y_train))
print('Accuracy Score:',accuracy_score(y_test,y_pred_lr))
print('Cross Validation Score:',np.mean(cross_val_score(lr,x_over,y_over,cv=5)))
print('Classification Report:',classification_report(y_test,y_pred_lr))
print('Confusion Matrix:',confusion_matrix(y_test,y_pred_lr))
```

### KNeighbors Classifier

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
print('Training score:',knn.score(x_train,y_train))
print('Accuracy Score:',accuracy_score(y_test,y_pred_knn))
print('Cross Validation Score:',np.mean(cross_val_score(knn,x_over,y_over,cv=5)))
print('Classification Report:',classification_report(y_test,y_pred_knn))
print('Confusion Matrix:',confusion_matrix(y_test,y_pred_knn))
```

### Decision Tree Classifier

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred_dt=dt.predict(x_test)
print('Training score:',dt.score(x_train,y_train))
print('Accuracy Score:',accuracy_score(y_test,y_pred_dt))
print('Cross Validation Score:',np.mean(cross_val_score(dt,x_over,y_over,cv=5)))
print('Classification Report:',classification_report(y_test,y_pred_dt))
print('Confusion Matrix:',confusion_matrix(y_test,y_pred_dt))
```

### Random Forest Classifier

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
print('Training score:',rf.score(x_train,y_train))
print('Accuracy Score:',accuracy_score(y_test,y_pred_rf))
print('Cross Validation Score:',np.mean(cross_val_score(rf,x_over,y_over,cv=5)))
print('Classification Report:',classification_report(y_test,y_pred_rf))
print('Confusion Matrix:',confusion_matrix(y_test,y_pred_rf))
```

### Hyper Parameter Tuning

### Logistic Regression

```python
#Hyperparameter tuning using GridSearchCV for Logistic Regression to find best parameters.
parameters={'penalty':['l1','l2']}
lr=LogisticRegression()
clf=GridSearchCV(lr,parameters,n_jobs=-1)
clf.fit(x_train,y_train)
print(clf.best_params_)
```

### KNeighbors Classifier

```python
#Hyperparameter tuning using GridSearchCV for Kneighborsclassifier to find best parameters.
parameters={ 'n_neighbors' :[1,5],'weights':['uniform','distance'],'algorithm':['kd_tree','brute']}
knn=KNeighborsClassifier()
clf=GridSearchCV(knn,parameters,n_jobs=-1)
clf.fit(x_train,y_train)
print(clf.best_params_)
```

### Decision Tree Classifier

```python
#Hyperparameter tuning using GridSearchCV for DecisionTreeClassifier to find out best parameters.
parameters={'criterion':['gini','entropy']}
dt=DecisionTreeClassifier()
clf=GridSearchCV(dt,parameters,n_jobs=-1)
clf.fit(x_train,y_train)
print(clf.best_params_)
```

### Random Forest Classifier

```python
#Hyperparameter tuning using GridSearchCV for RandomForestClassifier to find out best parameters.
parameters={'n_estimators' :[100,200,300],'criterion':['gini','entropy'],'max_depth':[10,15,20]}
rf=RandomForestClassifier()
clf=GridSearchCV(rf,parameters,n_jobs=-1)
clf.fit(x,y)
print(clf.best_params_)
```
